```python
df = df.dropna() 

df['Year'] = df['Year'].astype(int)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os


os.makedirs('/kaggle/working/data', exist_ok=True)
os.makedirs('/kaggle/working/figures', exist_ok=True)


processed_data_path = '/kaggle/working/data/processed_women_in_stem.csv'
df.to_csv(processed_data_path, index=False)
print(f"‚úÖ –î–∞–Ω–Ω—ã–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã: {processed_data_path}")


plt.figure(figsize=(12, 6))
ax = sns.countplot(data=df, x='STEM Fields', palette='viridis')


for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points')

plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –ø–æ STEM-–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è–º', pad=20)
plt.xlabel('–û–±–ª–∞—Å—Ç—å STEM')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π')
plt.xticks(rotation=45)


figure_path = '/kaggle/working/figures/stem_fields_distribution.png'
plt.savefig(figure_path, dpi=300, bbox_inches='tight')
print(f"‚úÖ –ì—Ä–∞—Ñ–∏–∫ —Å–æ—Ö—Ä–∞–Ω–µ–Ω: {figure_path}")
plt.show()


print("\nüìÇ –°–æ–¥–µ—Ä–∂–∏–º–æ–µ —Ä–∞–±–æ—á–µ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏:")
!ls -lh /kaggle/working/
print("\nüìÇ –°–æ–¥–µ—Ä–∂–∏–º–æ–µ –ø–∞–ø–∫–∏ data:")
!ls -lh /kaggle/working/data/
print("\nüìÇ –°–æ–¥–µ—Ä–∂–∏–º–æ–µ –ø–∞–ø–∫–∏ figures:")
!ls -lh /kaggle/working/figures/
```
